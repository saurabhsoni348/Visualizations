```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.Series([1, 3, 2, 5, 7, 8, 9, 10])
plt.figure(figsize=(10, 6))
data.plot(kind='line')
plt.title('Line Graph Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 12]
plt.figure(figsize=(10, 6))
plt.bar(categories, values)
plt.title('Bar Graph Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(10, 6))
plt.boxplot(data)
plt.title('Box Plot Example')
plt.xlabel('Dataset')
plt.ylabel('Value')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```
